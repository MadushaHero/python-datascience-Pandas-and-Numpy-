```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
exp_vals = [1400,600,300,410,250]
exp_labels = ["Home Rent","Food","Phone/Internet Bill","Car ","Other Utilities"]
plt.pie(exp_vals,labels=exp_labels)
```

```python
plt.pie(exp_vals,labels=exp_labels, shadow=True)
plt.axis("equal")
plt.show()
```

```python
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels, shadow=True,radius=1.5)

plt.show()
```

```python

plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels, autopct='%1.3f%%',shadow=True,radius=1.5)

plt.show()
```

```python
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels, autopct='%1.3f%%',shadow=True,radius=1.5,explode=[0,0.2,0,0,0])

plt.show()
```

```python
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels, autopct='%1.3f%%',shadow=True,radius=1.5,explode=[0,0.2,0,0,0],startangle=45)

plt.show()
```

```python
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels, autopct='%1.3f%%',shadow=True,radius=1.5,explode=[0,0.2,0,0,0],startangle=45,colors='r',textprops={'color':"w"},wedgeprops={"edgecolor":"g",'linewidth': 5, 'linestyle': 'dashed', 'antialiased': True})


plt.savefig("piechart.png")
```
